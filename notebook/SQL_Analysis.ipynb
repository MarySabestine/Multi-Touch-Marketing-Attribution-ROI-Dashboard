```python
#spend by channel (imp in sql interview questions)
query = """
SELECT
    channel,
    SUM(spend) AS total_spend
FROM marketing_spend
GROUP BY channel
ORDER BY total_spend DESC;
"""

pd.read_sql(query, conn)
```

```python
#campaign performance 
query = """
SELECT
    campaign,
    SUM(spend) AS total_spend
FROM marketing_spend
GROUP BY campaign
ORDER BY total_spend DESC;
"""

pd.read_sql(query, conn)
```

```python
#highest spend day
query = """
SELECT
    date,
    SUM(spend) AS daily_spend
FROM marketing_spend
GROUP BY date
ORDER BY daily_spend DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
#day with zero spend
query = """
SELECT *
FROM marketing_spend
WHERE spend = 0;
"""

pd.read_sql(query, conn)
```

```python
##windows functions
#Rank campaigns by total spend (This ranks campaigns based on total spend.)
query = """
SELECT
    campaign,
    SUM(spend) AS total_spend,
    RANK() OVER (
        ORDER BY SUM(spend) DESC
    ) AS campaign_rank
FROM marketing_spend
GROUP BY campaign;
"""

pd.read_sql(query, conn)
```

```python
#Row number within each channel(This assigns a row number to each record within its channel, ordered by spend.)
query = """
SELECT
    date,
    channel,
    campaign,
    spend,
    ROW_NUMBER() OVER (
        PARTITION BY channel
        ORDER BY spend DESC
    ) AS row_num
FROM marketing_spend;
"""

pd.read_sql(query, conn)
```

```python
#Previous day's spend (LAG) (This lets you compare each day's spend to the previous day's spend within the same channel.)
query = """
SELECT
    date,
    channel,
    spend,
    LAG(spend) OVER (
        PARTITION BY channel
        ORDER BY date
    ) AS previous_spend
FROM marketing_spend;
"""

pd.read_sql(query, conn)
```

```python
#Next day's spend (LEAD) (This returns the following day's spend for each channel.)
query = """
SELECT
    date,
    channel,
    spend,
    LEAD(spend) OVER (
        PARTITION BY channel
        ORDER BY date
    ) AS next_spend
FROM marketing_spend;
"""

pd.read_sql(query, conn)
```

```python
#First-Click Attribution
WITH_first_touch AS (
    SELECT
        user_id,
        channel,
        campaign,
        conversion,
        ROW_NUMBER() OVER (
            PARTITION BY user_id
            ORDER BY timestamp
        ) AS rn
    FROM customer_journey
)
SELECT
    user_id,
    channel,
    campaign
FROM first_touch
WHERE rn = 1;
```

```python
#First-Click Attribution (This gives the first marketing touchpoint for every user.)
query = """
WITH first_touch AS (
    SELECT
        user_id,
        channel,
        campaign,
        conversion,
        ROW_NUMBER() OVER (
            PARTITION BY user_id
            ORDER BY timestamp
        ) AS rn
    FROM customer_journey
)
SELECT
    user_id,
    channel,
    campaign
FROM first_touch
WHERE rn = 1;
"""

pd.read_sql(query, conn)
```

```python
#Last-Click Attribution (This gives the last touchpoint before the user's final interaction)
query = """
WITH last_touch AS (
    SELECT
        user_id,
        channel,
        campaign,
        conversion,
        ROW_NUMBER() OVER (
            PARTITION BY user_id
            ORDER BY timestamp DESC
        ) AS rn
    FROM customer_journey
)
SELECT
    user_id,
    channel,
    campaign
FROM last_touch
WHERE rn = 1;
"""

pd.read_sql(query, conn)
```

```python
#linear attribution (Each touchpoint receives an equal share of the conversion credit.)
query = """
SELECT
    user_id,
    timestamp,
    channel,
    campaign,
    ROUND(
        1.0 / COUNT(*) OVER (PARTITION BY user_id),
        2
    ) AS attribution_credit
FROM customer_journey
ORDER BY user_id, timestamp;
"""

pd.read_sql(query, conn)
```
