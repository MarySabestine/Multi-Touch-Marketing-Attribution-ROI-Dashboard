```python
#import libraries
import pandas as pd
import sqlite3
```

```python
#connect to sqlite 
conn = sqlite3.connect("../database/marketing.db")

print("Database Connected!")
```

```python
#loading the clean csv
df = pd.read_csv("../data/cleaned/ad_spend_cleaned.csv")

df.head()
```

```python
#create the sql table
df.to_sql(
    "marketing_spend",
    conn,
    if_exists="replace",
    index=False
)

print("Table Created Successfully!")

```
