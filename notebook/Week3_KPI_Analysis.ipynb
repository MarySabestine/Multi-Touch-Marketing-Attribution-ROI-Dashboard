```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/cleaned/ad_spend_cleaned.csv")
```

```python
df.head()
```

```python
np.random.seed(42)

df["clicks"] = np.random.randint(50, 500, len(df))
df["conversions"] = np.random.randint(5, 50, len(df))
df["revenue"] = np.random.randint(5000, 50000, len(df))
```

```python
df.head()
```

```python
#Calculate KPIs
#CPC (Cost Per Click)
df["CPC"] = df["spend"] / df["clicks"]
```

```python
#CAC (Customer Acquisition Cost)
df["CAC"] = df["spend"] / df["conversions"]
```

```python
#ROAS (Return on Ad Spend)
df["ROAS"] = df["revenue"] / df["spend"]
```

```python
#View Results
df[
    [
        "channel",
        "campaign",
        "spend",
        "clicks",
        "conversions",
        "revenue",
        "CPC",
        "CAC",
        "ROAS",
    ]
].head()
```

```python
#KPI Summary by Channel
summary = (
    df.groupby("channel")
      .agg({
          "spend": "sum",
          "clicks": "sum",
          "conversions": "sum",
          "revenue": "sum"
      })
)

summary["CPC"] = summary["spend"] / summary["clicks"]
summary["CAC"] = summary["spend"] / summary["conversions"]
summary["ROAS"] = summary["revenue"] / summary["spend"]

summary
```

```python
#Create Dimension Tables
#Dimension: Channel

dim_channel = df[["channel"]].drop_duplicates().reset_index(drop=True)

dim_channel["channel_id"] = range(1, len(dim_channel) + 1)

dim_channel = dim_channel[["channel_id", "channel"]]

dim_channel
```

```python
#Dimension: Campaign
dim_campaign = df[["campaign"]].drop_duplicates().reset_index(drop=True)

dim_campaign["campaign_id"] = range(1, len(dim_campaign) + 1)

dim_campaign = dim_campaign[["campaign_id", "campaign"]]

dim_campaign
```

```python
#Dimension: Date
dim_date = pd.DataFrame()

dim_date["date"] = df["date"].drop_duplicates().sort_values().reset_index(drop=True)

dim_date["year"] = dim_date["date"].dt.year
dim_date["month"] = dim_date["date"].dt.month
dim_date["month_name"] = dim_date["date"].dt.month_name()
dim_date["day"] = dim_date["date"].dt.day
dim_date["weekday"] = dim_date["date"].dt.day_name()
dim_date["quarter"] = dim_date["date"].dt.quarter

dim_date.head()
```

```python
dim_date = pd.DataFrame()

dim_date["date"] = pd.to_datetime(
    df["date"].drop_duplicates().sort_values().reset_index(drop=True)
)

dim_date["year"] = dim_date["date"].dt.year
dim_date["month"] = dim_date["date"].dt.month
dim_date["month_name"] = dim_date["date"].dt.month_name()
dim_date["day"] = dim_date["date"].dt.day
dim_date["weekday"] = dim_date["date"].dt.day_name()
dim_date["quarter"] = dim_date["date"].dt.quarter

dim_date.head()
```

```python
#Create the Fact Table
#Merge the IDs into the main data:

fact_marketing = df.merge(dim_channel, on="channel")
fact_marketing = fact_marketing.merge(dim_campaign, on="campaign")
```

```python
#Now select the required columns:
fact_marketing = fact_marketing[
    [
        "date",
        "channel_id",
        "campaign_id",
        "spend",
        "clicks",
        "conversions",
        "revenue",
        "CPC",
        "CAC",
        "ROAS",
    ]
]

fact_marketing.head()
```

```python
dim_channel.to_csv("../data/processed/Dim_Channel.csv", index=False)

dim_campaign.to_csv("../data/processed/Dim_Campaign.csv", index=False)

dim_date.to_csv("../data/processed/Dim_Date.csv", index=False)

fact_marketing.to_csv("../data/processed/Fact_Marketing.csv", index=False)

print("Star Schema tables saved successfully!")
```
