```python
def explore_dataset(df, name):
   
    print("=" * 60)
    print(f"Dataset: {name}")
    print("=" * 60)

    # Shape
    print(f"\nShape: {df.shape[0]} rows × {df.shape[1]} columns")

    # Column names
    print("\nColumns:")
    print(df.columns.tolist())

    # Data types and missing values
    print("\nDataset Information:")
    print(df.info())

    # Missing values
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Summary statistics
    print("\nSummary Statistics:")
    print(df.describe(include='all').T)

    # First five rows
    print("\nFirst Five Rows:")
    display(df.head())
```

Handling Missing UTM Parameters

```python


def missing_and_unique(df, dataset_name):
    """
    Display missing values and unique values for each column.

    Parameters:
        df (DataFrame): Dataset to analyze.
        dataset_name (str): Name of the dataset.
    """

    print("=" * 70)
    print(f"Dataset: {dataset_name}")
    print("=" * 70)

    summary = pd.DataFrame({
        "Data Type": df.dtypes,
        "Missing Values": df.isnull().sum(),
        "Missing (%)": round((df.isnull().sum() / len(df)) * 100, 2),
        "Unique Values": df.nunique()
    })

    display(summary)

    print("\nUnique Values by Column")
    print("-" * 70)

    for col in df.columns:
        print(f"\n{col}:")
        unique_vals = df[col].dropna().unique()

        # Display only if there aren't too many unique values
        if len(unique_vals) <= 20:
            print(unique_vals)
        else:
            print(f"{len(unique_vals)} unique values (too many to display)")
```

Standardizing Timestamp format

```python
def convert_to_datetime(df, columns, dayfirst=False):
    """
    Convert one or more columns to datetime format.

    Parameters:
        df (DataFrame): Input DataFrame.
        columns (list): List of column names to convert.
        dayfirst (bool): Set to True if dates are in DD/MM/YYYY format.

    Returns:
        DataFrame: DataFrame with converted datetime columns.
    """

    df = df.copy()

    for col in columns:
        if col in df.columns:
            df[col] = pd.to_datetime(
                df[col],
                errors="coerce",
                dayfirst=dayfirst
            )
            print(f"✓ '{col}' converted to datetime.")
        else:
            print(f"✗ Column '{col}' not found.")

    return df
```

Inspect Datetime

```python
def inspect_datetime(df, columns):
    """
    Inspect datetime columns in a DataFrame.

    Parameters:
        df (DataFrame): Input DataFrame.
        columns (list): List of datetime columns to inspect.
    """

    for col in columns:
        print("=" * 70)
        print(f"Datetime Inspection: {col}")
        print("=" * 70)

        if col not in df.columns:
            print(f"❌ Column '{col}' not found.\n")
            continue

        print(f"Data Type: {df[col].dtype}")
        print(f"Missing (NaT): {df[col].isna().sum()}")
        print(f"Unique Values: {df[col].nunique()}")

        if pd.api.types.is_datetime64_any_dtype(df[col]):
            print(f"Earliest Date: {df[col].min()}")
            print(f"Latest Date: {df[col].max()}")

        print("\nFirst 5 Values:")
        print(df[col].head())

        print("\nLast 5 Values:")
        print(df[col].tail())

        print("\n")
```

Converting to UTC

```python
def convert_to_utc(df, columns):

    df = df.copy()

    for col in columns:
        if col not in df.columns:
            print(f"❌ Column '{col}' not found.")
            continue

        try:
            # Convert to datetime and standardize to UTC
            df[col] = pd.to_datetime(
                df[col],
                errors="coerce",
                utc=True
            )

            print(f"✓ '{col}' converted to UTC.")

        except Exception as e:
            print(f"❌ Error converting '{col}': {e}")

    return df
```

```python
def create_date_components(df, columns):

    df = df.copy()

    for col in columns:
        if col not in df.columns:
            print(f"❌ Column '{col}' not found.")
            continue

        if not pd.api.types.is_datetime64_any_dtype(df[col]):
            print(f"❌ '{col}' is not a datetime column.")
            continue

        # Create date components
        df[f"{col}_quarter"] = df[col].dt.quarter
        df[f"{col}_month_name"] = df[col].dt.month_name()
        df[f"{col}_day_name"] = df[col].dt.day_name()
        df[f"{col}_day_of_week"] = df[col].dt.dayofweek
        df[f"{col}_week"] = df[col].dt.isocalendar().week
        df[f"{col}_hour"] = df[col].dt.hour
        df[f"{col}_is_weekend"] = df[col].dt.dayofweek >= 5

        print(f"✓ Date components created for '{col}'.")

    return df
```

Inspecting Duplicates

```python
def check_duplicates(df, dataset_name, remove=False):

    print("=" * 70)
    print(f"Duplicate Check: {dataset_name}")
    print("=" * 70)

    duplicate_count = df.duplicated().sum()

    print(f"Total Rows: {len(df)}")
    print(f"Duplicate Rows: {duplicate_count}")

    if duplicate_count > 0:
        print("\nDuplicate Records:")
        display(df[df.duplicated()])

        if remove:
            df = df.drop_duplicates().reset_index(drop=True)
            print(f"\n✓ Removed {duplicate_count} duplicate rows.")
            print(f"New Shape: {df.shape}")
    else:
        print("✓ No duplicate rows found.")

    return df
```

User ID Standardization

```python
def standardize_user_id(df, column="user_id"):

    df = df.copy()

    if column not in df.columns:
        print(f"❌ Column '{column}' not found.")
        return df

    # Convert to string
    df[column] = df[column].astype(str)

    # Remove leading/trailing spaces
    df[column] = df[column].str.strip()

    # Remove spaces within IDs
    df[column] = df[column].str.replace(" ", "", regex=False)

    # Convert to uppercase
    df[column] = df[column].str.upper()

    # Remove special characters (keep only letters and numbers)
    df[column] = df[column].str.replace(r"[^A-Z0-9]", "", regex=True)

    # Replace string representations of missing values with NA
    df[column] = df[column].replace(
        ["NAN", "NONE", "NULL", ""],
        pd.NA
    )

    print(f"✓ '{column}' has been standardized.")

    return df
```

```python
Customer_Table = standardize_user_id(Customer_Table)

User_Weblogs = standardize_user_id(User_Weblogs)

Conversion_Table = standardize_user_id(Conversion_Table)
```

```python
def standardize_channel(df, column="channel"):
    
    df = df.copy()

    if column not in df.columns:
        print(f"❌ Column '{column}' not found.")
        return df

    # Clean text
    df[column] = (
        df[column]
        .astype(str)
        .str.strip()
        .str.lower()
    )

    # Mapping dictionary
    channel_map = {
        # Facebook / Meta
        "facebook": "Facebook",
        "fb": "Facebook",
        "meta": "Facebook",

        # Google Ads
        "google": "Google Ads",
        "google ads": "Google Ads",
        "googleads": "Google Ads",
        "gads": "Google Ads",

        # Email
        "email": "Email",
        "e-mail": "Email",
        "mail": "Email",

        # Organic
        "organic": "Organic",
        "organic search": "Organic",
        "seo": "Organic",

        # Direct
        "direct": "Direct",

        # LinkedIn
        "linkedin": "LinkedIn",
        "linkedin ads": "LinkedIn",

        # Instagram
        "instagram": "Instagram",
        "ig": "Instagram",

        # Twitter/X
        "twitter": "Twitter/X",
        "x": "Twitter/X",

        # YouTube
        "youtube": "YouTube",

        # Display
        "display": "Display",
        "display ads": "Display",

        # Referral
        "referral": "Referral"
    }

    # Apply mapping
    df[column] = df[column].replace(channel_map)

    # Capitalize any values not in the mapping
    df[column] = df[column].str.title()

    print(f"✓ '{column}' has been standardized.")

    return df
```

```python
Adspend_Table = standardize_channel(Adspend_Table)

User_Weblogs = standardize_channel(User_Weblogs)

Customer_Table = standardize_channel(Customer_Table)
```

```python
def save_to_csv(df, filename, folder="../data/cleaned_data", index=False):
    
    # Create the folder if it doesn't exist
    os.makedirs(folder, exist_ok=True)

    # Create the full file path
    file_path = os.path.join(folder, filename)

    # Save the DataFrame
    df.to_csv(file_path, index=index)

    print(f"✓ Dataset successfully saved to:\n{file_path}")
```

```python
save_to_csv(Adspend_Table, "cleaned_Adspend_Table.csv")

save_to_csv(User_Weblogs, "cleaned_User_Weblogs.csv")

save_to_csv(Customer_Table, "cleaned_Customer_Table.csv")

save_to_csv(Conversion_Table, "cleaned_Conversion_Table.csv")
```
