```python
#Imports required libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Using Pandas library to read the csv file.This stores CSV file in ram.Dynamically stored.
customer=pd.read_csv("/workspaces/Multi-Touch-Marketing-Attribution-ROI-Dashboard/Raw_Data/dim_customer.csv.xls")
```

```python
#Describe() functions gives a statistical Summary of the dataframe.
customer.describe()
```

```python
#Converting customer['first_touch_date'] column from str datatype to datetime dataype.
customer['first_touch_date'] = pd.to_datetime(customer['first_touch_date'])

#Normalizing datetime format.Converts time in datetime to zeros.
#Handles data efficiently while working with Power BI.
customer['first_touch_date'] = customer['first_touch_date'].dt.normalize()
```

```python
#Converting customer['last_touch_date'] column from str datatype to datetime dataype.
customer['last_touch_date'] = pd.to_datetime(customer['last_touch_date'])

#Normalizing datetime format.Converts time in datetime to zeros.
#Handles data efficiently while working with Power BI.
customer['last_touch_date'] = customer['last_touch_date'].dt.normalize()
```

```python
#Converting customer['conversion_date'] column from str datatype to datetime dataype.
customer['conversion_date'] = pd.to_datetime(customer['conversion_date'])

#Normalizing datetime format.Converts time in datetime to zeros.
#Handles data efficiently while working with Power BI.
customer['conversion_date'] = customer['conversion_date'].dt.normalize()
```

```python
print(customer["region"].value_counts().sort_index())
print(f"Total Values :   {customer["region"].value_counts().sum()}")
plt.hist(customer["region"])
plt.xlabel("Region")
plt.title("Distribution of Region")
```

```python
print(customer["total_touchpoints"].value_counts().sort_index())
plt.hist(customer["total_touchpoints"], bins=16)
plt.xlabel("Total Touchpoints")
plt.ylabel("Number of Customers")
plt.title("Distribution of Customer Touchpoints")
plt.show()
```

```python
customer["acquisition_channel"].value_counts().plot(kind="bar")
plt.xlabel("Channel")
plt.ylabel("Customers")
plt.title("Customers by Acquisition Channel")
plt.show()
```

```python
customer["conversion_flag"].value_counts().plot(kind="bar")
plt.xlabel("Conversion")
plt.ylabel("Count")
plt.title("Converted vs Not Converted")
plt.show()
```
