```python
#Imports required libraries.
import pandas as pd
```

```python
#Using Pandas library to read the csv file.This stores CSV file in ram.Dynamically stored.
conversions=pd.read_csv("conversions.csv.xls")
```

```python
#head() function is used to see first few rows of csv file.
conversions.head()
```

```python
#Similar to head() , but retreives a random sample of rows for examining. 
conversions.sample(5)
```

```python
#Using info() function helps understand data
#Can verify Columns, total columns, Non-Null Columns, Datatypes
conversions.info()
```

```python
#Describe() functions gives a statistical Summary of the dataframe.
conversions.describe()
```

```python
#Converting Conversion_date column from str datatype to datetime dataype.
conversions['conversion_date'] = pd.to_datetime(conversions['conversion_date'])

#Normalizing datetime format.Converts time in datetime to zeros.
#Handles data efficiently while working with Power BI.
conversions['conversion_date'] = conversions['conversion_date'].dt.normalize()
```

```python
#Checking Duplicates.
print("Duplicate rows:", conversions.duplicated().sum())
```

```python
#Checking missing values.
print(conversions.isnull().sum())
```
